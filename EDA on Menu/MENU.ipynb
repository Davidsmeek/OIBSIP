```python
import pandas as pd
df = pd.read_csv(r'\Users\User\Desktop\INTERSHIP WORK\menu (1).csv')
```

```python
%whos
df.head()
```

```python
# Inspect basic structure
print(df.head())
print(df.info())

```

```python
# Clean column names (optional)
df.columns = df.columns.str.strip()

```

```python
# Check for and handle missing values
df = df.dropna()

```

```python
# Verify category distribution
print(df['Category'].value_counts())
```

```python
# Basic statistics for numeric columns
desc = df.describe()
print(desc)

# Most common items and averages
most_common_item = df['Item'].mode()[0]
avg_calories = df['Calories'].mean()
print(f"Most common item: {most_common_item}, Average calories: {avg_calories:.2f}")
```

```python
# Average calories per category
category_avg = df.groupby('Category')['Calories'].mean().sort_values(ascending=False)
print(category_avg)
```

```python
# Top 5 highest protein items
top_protein = df[['Item', 'Category', 'Protein']].sort_values(by='Protein', ascending=False).head(5)
print("Top 5 Protein-Rich Items:\n", top_protein)

# Items with highest fat content
top_fat = df[['Item', 'Total Fat']].sort_values(by='Total Fat', ascending=False).head(5)
print("Most Fat-Dense Items:\n", top_fat)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot: Average calories by category
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Category', y='Calories', estimator='mean', ci=None)
plt.xticks(rotation=45)
plt.title('Average Calories per Category')
plt.tight_layout()
plt.show()

# Heatmap of correlation between nutrients
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Nutritional Correlation Heatmap')
plt.tight_layout()
plt.show()
```
